4장 기본 그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40])
# plt.show() print 안해도 출력됨 show 하지 않아도 그래프가 그려짐
```

```python
%mapplotlib inline # 그림을 이 문서 안에 그려라
```

```python
plt.plot([1,2,3,4],[12,43,25,15,24]) #x,y 축 데이터의 개수가 맞지 않으면 에러가 발생
```

```python
plt.title('Graph title')          # 제목 작성
plt.plot([1,2,3,4],[12,43,25,15], label='test1', color='blue', linestyle='--') #(x, y) 
plt.plot([1,2,3,4],[20,44,55,10], label='test2', color='red', ls=':') #(x, y) 
plt.plot([1,2,3,4],[1,5,10,30], label='test3', color='black') #(x, y) 
plt.legend(loc=2)                      # 범례를 그려라 P61 4=오른쪽 하단, 2=왼쪽상단
```

```python
plt.title('Graph title')          # 제목 작성
plt.plot([1,2,3,4],[12,43,25,15], 'r.', label='test1') #마커 모양 변경 r.
plt.plot([1,2,3,4],[20,44,55,10], 'g^',label='test2') #마커 모양 변경 g^
plt.plot([1,2,3,4],[1,5,10,30], label='test3', color='black') #(x, y) 
plt.legend(loc=2)                      # 범례를 그려라 P61 4=오른쪽 하단, 2=왼쪽상단
```
