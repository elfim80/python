```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 30세 이하의 인구 비율이 가장 높은 도시 5개를 찾고, 도시 이름과 인구수를 그래프로 나타내세요.
df = pd.read_csv('age.csv', index_col=0)
df = df.dropna() # 누락 행 제거
df = df.replace(',','', regex=True)
a = df.astype('int')
```

```python
삼십이하총인구수 = np.sum(a.iloc[:,2:33], axis=1) # 삼십이하 인구수를 다 더함
b = 삼십이하총인구수.div(a.iloc[:,0], axis=0)    # 삼십이하 총인구수를 총인구수로 나눔.
i = b.sort_values(ascending=False)[:5]           # 큰 것 5개
```

```python
plt.rc('font', family='Malgun Gothic')
p = plt.title('30세 이하의 인구 비율이 가장 높은 도시 5개')
p = plt.barh(i.index, i)
```

```python
# 4. 유임 승차에 비해 무임 승차 인원의 비율이 높은 지하철역 5개의 이름과 비율을 그래프로 나타내세요.
df = pd.read_excel('2019년 05월  교통카드 통계자료.xls',
                  sheet_name=2, encording='cp949', skiprows=0, index_col=3)
subway = df.iloc[:,3:-1]
subway = subway.replace(',','',regex=True).astype('int')
```

```python
유임승차 = subway['유임승차']
무임승차 = subway['무임승차']
비율 = 무임승차/유임승차
subway['비율'] = 비율
```

```python
subway.head(4)
```

```python
y = 비율.sort_values(ascending=False)[:5]
p = plt.title('유임 승차에 비해 무임 승차 인원의 비율이 높은 지하철역 5개')
p = plt.bar(y.index, y)
```
