```python
# unit 7 p96~p108

import csv

f = open('age.csv', encoding='UTF-8')  # 파일 여는 내장 함수
data = csv.reader(f)   
result = []                            # 빈 리스트 만들기

for row in data :
    if '신도림' in row[0] :           # '신도림'이 포함된 행정구역 찾기
        for i in row[3:] :            # 0세부터 끝(100세 이상)까지 모든 연령에 대해 반복하기
            result.append(int(i))     # 해당 연령의 인구수 리스트에 순서대로 저장하기
print(result)                         # 0세부터 100세 이상까지의 인구수 출력하기
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')             # 격자 무늬 스타일 지정
plt.plot(result)
plt.show()
```

```python
# unit 8 p109~p110

import csv

f = open('age.csv', encoding='UTF-8')  # 파일 여는 내장 함수
data = csv.reader(f)   
result = []                            # 빈 리스트 만들기
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')

for row in data :
    if name in row[0] :               # 입력받은 지역명이 포함된 행정구역 찾기
        for i in row[3:] :            # 0세부터 끝(100세 이상)까지 모든 연령에 대해 반복하기
            result.append(int(i))     # 해당 연령의 인구수 리스트에 순서대로 저장하기

print(result)                         # 0세부터 100세 이상까지의 인구수 출력하기        
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')                      # 격자 무늬 스타일 지정
plt.rc('font', family='Malgun Gothic')       #폰트는 맑은 고딕
plt.title(name+' 지역의 인구 구조')          # 그래프의 제목
plt.plot(result)
plt.show()
```
