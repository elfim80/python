```python
df = df.dropna() # 누락 행 제거
df = df.replace(',','', regex=True)
```

```python
a = df.astype('int')
a
```

```python
a.info
a.dtypes
```

```python
# 인구를 총 인구수로 나눈다. : 인구수 -> 비율로 변경됨
#a.div(a['2019년02월_계_총인구수'], axis=0)
b = a.div(a.iloc[:,0], axis=0) # 비율 데이터
b
```

```python
c = b.iloc[:,2:]
c
```

```python
# P270
# 궁금한 지역의 이름을 입력받음
name = input('궁금한 지역이 어디인가요?')
```

```python
c.index.str.contains(name).sum()
mask = c.index.str.contains(name)
df2 = c[mask]
df2.head(2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family='Malgun Gothic')
m = df2.T.plot()
```

```python
# P271
# 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 시각화

df = c
x = df.sub(df2.iloc[0], axis=1) # 원본과 내가 관심있는 지역의 0번째 행을 뺀다
#y = np.power(x, 2) # 제곱
y = np.abs(x)  # 절대값
z = y.sum(axis=1) # 나이별 차이를 다 누적시킨다.
i = z.sort_values().index[:5]
```

```python
m = df.loc[i].T.plot()
```
